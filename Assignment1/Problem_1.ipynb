```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# This part is for reading the file
f = open("test.txt")
test_str = f.read().lower()
f.close()
```

```python
# Setups
char_list = list(test_str)
char_dict = {}
char_dict_sorted = {}
char_num = 0
```

```python
for key in char_list:
    if key.isalpha():
        char_num += 1
        if char_dict.get(key) == None:
            char_dict[key] = 1
        else:
            char_dict[key] += 1
            
key_list  = sorted(char_dict.keys())
for key in key_list:
    char_dict_sorted[key] = char_dict[key]
print("Below is the histogram as dictionary")
print(char_dict_sorted)
print("------------------------------------")

np_char_dict = np.array(char_dict_sorted.items())
print("Below is the histogram as Numpy array")
print(np_char_dict)
print("------------------------------------")
print(char_num)
```

```python
from matplotlib.ticker import FuncFormatter
norm_freq = {}
for i in char_dict_sorted.keys():
    norm_freq[i]=char_dict_sorted[i]/char_num
def to_percent(y,position):
    return str(100*y)+"%"
plt.title('Bar graph on normalized frequencies(Dictionary)',fontsize='large', fontweight='bold')
plt.xlabel('letters')
plt.ylabel('normalized frequencies')
fomatter=FuncFormatter(to_percent)
plt.gca().yaxis.set_major_formatter(fomatter)
plt.bar(norm_freq.keys(),norm_freq.values())
```

```python
plt.title('Bar graph on unnormalized frequencies(Dictionary)',fontsize='large', fontweight='bold')
plt.xlabel('letters')
plt.ylabel('unnormalized frequencies')
plt.bar(char_dict_sorted.keys(),char_dict_sorted.values())
```
